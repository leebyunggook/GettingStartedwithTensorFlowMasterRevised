# Linear Regression

```python
import numpy as np

number_of_points = 200
x_point = []
y_point = []
a = 0.22
b = 0.78
for i in range(number_of_points):
    x = np.random.normal(0.0,0.5)
    y = a*x + b + np.random.normal(0.0,0.1)
    x_point.append([x])
    y_point.append([y])
```

```python
import matplotlib.pyplot as plt

plt.plot(x_point, y_point, 'ro', label='Input Data')
plt.legend()
plt.show()
```

```python
import tensorflow as tf

A = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
B = tf.Variable(tf.zeros([1]))
y = A * x_point + B

cost_function = tf.reduce_mean(tf.square(y - y_point))
optimizer = tf.train.GradientDescentOptimizer(0.5)
train = optimizer.minimize(cost_function)

model = tf.global_variables_initializer()

with tf.Session() as session:
    session.run(model)
    for step in range(0,41):
        session.run(train)
        if (step % 5) == 0:
            plt.plot(x_point, y_point, 'o', label='step = {}'.format(step))
            plt.plot(x_point, session.run(A) * x_point + session.run(B))
            plt.legend()
            plt.show()
    print(session.run(A), session.run(B))
```

# Quadratic Regression

```python
number_of_points = 40
x_point = []
y_point = []
a = -0.72
b = 0.78
c = 1.2
for i in range(number_of_points):
    x = np.random.normal(0.0,0.5)
    y = a*x*x + b*x + c + np.random.normal(0.0,0.1)
    x_point.append([x])
    y_point.append([y])
```

```python
_x = np.arange(-1,2.1,0.1)
#help(np.arange)
_y = a*_x*_x + b*_x + c
print(_x)
print(_y)
```

```python
import matplotlib.pyplot as plt

plt.plot(x_point, y_point, 'o', label='Input Data')
plt.plot(_x, _y)
plt.legend()
plt.show()
```

```python
A = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
B = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
C = tf.Variable(tf.zeros([1]))
y = A * x_point * x_point + B * x_point + C

cost_function = tf.reduce_mean(tf.square(y - y_point))
optimizer = tf.train.GradientDescentOptimizer(0.5)
train = optimizer.minimize(cost_function)

model = tf.global_variables_initializer()

with tf.Session() as session:
    session.run(model)
    for step in range(0,41):
        session.run(train)
        if (step % 5) == 0:
            plt.plot(x_point, y_point, 'o', label='step = {}'.format(step))
            plt.plot(_x, session.run(A)*_x*_x+session.run(B)*_x+session.run(C))
            plt.legend()
            plt.show()
    print(session.run(A), session.run(B), session.run(C))
print(a, b, c)    
```
